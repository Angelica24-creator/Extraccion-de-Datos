```python
#importación de librerias 

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt 
import seaborn as sns
```

```python
#cargar archivo sin indice 

df= pd.read_excel('Detalle precios y productos fabricados 2022.xlsx')
df 
```

```python
#informacion de las columnas de la base de datos 

df.info()
```

**Filtro por objeto**

Descripción: Apliqué el filtro por objeto a la columna NOMBRE_CLIENTE para que solo muestre las filas donde el nombre del cliente sea MOLINERA DE MEXICO

```python
#Filtro por objeto
filtro1=df[(df["NOMBRE_CLIENTE"] == "MOLINERA DE MEXICO")]
filtro1
```

```python
#Convertir archivo filtrado a csv

filtro1.to_csv("Filtro_objeto.csv")
```

**Filtro por filas**

Descripción: En el filtro por filas consecutivas quiero que me muestre los datos de las filas de la 100 a la 200 y en las filas no consecutivas solo quiero ver la fila 100, la fila 200 y la fila 400.


```python
#filtro2=df.iloc[100:201,:] #filas consecutivas
filtro2=df.iloc[[100,200,400],:] #filas no consecutivas 
filtro2
```

```python
filtro2.to_csv("Filtro_filas_NO_consecutivas.csv")
```

**Filtro por columnas**

```python
filtro3=df.iloc[:, 0:4]      #Columnas consecutivas 
#filtro3=df.iloc[:,[2,3,13]]  #Columnas no consecutivas 
filtro3
```

**Filtro Y**

Descrpción: Me va a mostrar los datos donde el precio unitario sea mayor a $1500 y el costo total calculado sea menor a $5000

```python
filtro8=df[(df["PRECIO_UNITARIO"]>1500) & (df["COSTO_TOTAL_CALCULADO"] < 5000)]
filtro8
```

```python
filtro8.to_csv("Filtro_Y.csv")
```

**Filtro O**

Descripción: Muestra solo los registros con fecha del 24 de enero o donde donde la cantidad fue mayor 20

```python
filtro9=df[(df["FECHA_DOC"] == "24/01/2022") | (df["CANT"] > 20)]
filtro9
```

```python
filtro9.to_csv("Filtro_O.csv")
```

**Filtro not**

Descripción: Me va a devolver todos los datos a excepción de los registros donde la vendedora fue Leticia.

```python
filtro10=df[~(df["NOMBRE_VENDEDOR"] == "LETICIA RAMIREZ HERNANDEZ")]
filtro10
```
