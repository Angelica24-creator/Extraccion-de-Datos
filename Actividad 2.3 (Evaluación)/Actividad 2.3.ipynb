```python
import pandas as pd
import numpy as np 
import matplotlib.pyplot as plt 
import seaborn as sns
```

```python
df = pd.read_csv("Rio de Janeiro.csv", encoding="latin1")
df
```

```python
df.info()
```

**Registros de 5 host name diferentes**

Primero ocupe el filtro de columnas para ver los datos de los host_name, despues utilice el filtro O para que solo me mostrara los registros de los host_name que queria, en este caso fueron: Erika, Nilda, Jaqueline, Robert y Celina.

```python
filtro1=df.iloc[:,11]
filtro1
```

```python
filtro2=df[(df["host_name"] == "Erika") | (df["host_name"] == "Nilda") | (df["host_name"] == "Jaqueline") | (df["host_name"] == "Robert") | (df["host_name"] == "Celina")]
filtro2
```

```python
filtro2.to_csv("Filtro O_5 HOST NAME DIFERENTES.csv")
```

**Host que se hallan unido antes del año 2019**

Primero la columna host_since la cambie a tipo datatime para tener más seguridad de comparar fechas, despues ocupe el filtro por comparación donde pido que la columna host_since solo tome los registros con fechas antes o iguales al 31 de diciembre de 2018 para que las fechas sean antes del año 2019.

```python
#df['host_since'] = pd.to_datetime(df['host_since'], errors='coerce')
#df.info()

filtro3=df[df["host_since"] <= "2018-12-31"]
filtro3
```

```python
filtro2.to_csv("Filtro Comparacion Host antes de 2019.csv")
```

**Registros de los host que responden a más tardar en 1 día**

Ocupe un filtro NOT para que no me salieran los registros donde tuviera en la columna host_response_time la leyenda de "a few days or more" que significa que tarda en responder un día o más por lo tanto los demás registros están correctos.


```python
filtro3=df[~(df["host_response_time"] == "a few days or more")]
filtro3
```

```python
filtro3.to_csv("Filtro NOT Host que responden a más tardar en 1 día.csv")
```

**Los registros que tengan al menos 5 accommodates**

Nuevamente ocupe el filtro de comparacion para que me comparara la columna accommodates con valores de registros menores o iguales a 5

```python
filtro4=df[df["accommodates"] <= 5]
filtro4
```

```python
filtro4.to_csv("Filtro Comparacion Al menos 5 accommodates.csv")
```

**Los registros de los tipos de cuarto: "Entire home" y "Private room"**

El filtro O para que me de una opción o la otra, ya que no es necesario tener ambos y solo lo aplica a la comlumna "room_type"

```python
filtro5= df[(df["room_type"]=="Private room") | (df["room_type"]=="Entire home/apt")]
filtro5
```

```python
filtro5.to_csv("Filtro O Registros de los tipos de cuarto.csv")
```

**Los registros que "a lo mucho cuenten con 4 camas"**

Use el filtro por comparación, para que comparara los registros de la columna "beds" con el valos 4, 3, 2, 1 e incluso 0.

```python
filtro6=df[df["beds"] <= 4]
filtro6
```

```python
filtro6.to_csv("Filtro Comparacion A lo mucho 4 camas.csv")
```

**Los registros que "tengan 1 baño o que sean superhost" pero con la condición de que "todos sean instant_bookable"**

Se ocupo el filtro O para la primera parte, que tenga el registro o solo un baño o que tenga superhost y despues el filtro Y para que una vez filtrado eso lo agregue solo si tiene el instant bookable.

```python
filtro7=df[(df["bathrooms"] == 1) | (df["host_is_superhost"] == "t") & (df["instant_bookable"] == "t")]
filtro7
```

```python
filtro7.to_csv("Filtro O_Y Baños superhost.csv")
```

**Los registros cuyo "review_scores_cleanliness sea menor de 3 y review_scores_ rating sea mayor de 3"**

El filtro Y para que tome al mismo tiempo los valores menores de 3 y los valores mayores a 3 en sus repectivas columnas.

```python
filtro8=df[(df["review_scores_cleanliness"]< 3) & (df["review_scores_rating"] > 3)]
filtro8
```

```python
filtro8.to_csv("Filtro Y Baños review_scores_cleanliness y review_scores_ rating.csv")
```
